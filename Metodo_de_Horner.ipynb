# Método de Horner:

```python
from math import *
import numpy as np
import matplotlib . pyplot as plt
```

```python
def f(x):
    return 2*x**4 - 3*x**2 + 3*x - 4
```

```python
x0 = -2   # Parámetro inicial
tolerancia = 0.00001   # Tolerancia para el método de Newton
max_iter = 100   # Número máximo de iteraciones
```

```python
# Método de Horner:

coeficientes = [2, 0, -3, 3, -4]  # Creamos una lista con los coeficientes del polinomio

for i in range(0, max_iter + 1):    # Creamos un ciclo que repita el mismo proceso hasta el número máximo de iteraciones
    p = coeficientes[0]  # Inicializamos p con el primer coeficiente
    derivada_p = 0         # Inicializamos la derivada con 0

    # Método de Horner para evaluar el polinomio y su derivada
    for j in range(1, len(coeficientes)):
        derivada_p = derivada_p * x0 + p   # Derivada en x0
        p = p * x0 + coeficientes[j]  # Polinomio en x0

    # Ver si la derivada es cero (ya que esto haría fallar el método de Newton)
    if derivada_p == 0:
        print("La derivada es cero. El método de Newton falla.")
        break

    x1 = x0 - p / derivada_p   # Actualizamos el valor de x1 con el método de Newton

    error = abs(x1 - x0)   # Calculamos el error

    # Imprimir los resultados de la iteración
    print("Iteración {}: x = {:.6f}, Error = {:.6f}".format(i + 1, x1, error))

    # Verificamos si el error es menor que la tolerancia, en caso de serlo terminamos el código
    if error < tolerancia:
        print("\nLa solución aproximada es x = {:.6f}".format(x1))
        break

    # Actualizamos el valor de x0 para la siguiente iteración
    x0 = x1
```

```python
# Valores del eje x que toma el gráfico.
x=np.linspace(-50, 50,400)
```

```python
# Gráfica de la función
plt.figure(figsize=(5,5))
plt.title(r'Gráfica de $2x^4 - 3x^2 + 3x - 4$', fontsize=20)
plt.xlabel('Eje x', fontsize=14)
plt.ylabel('Eje y', fontsize=14)
plt.plot(x, f(x), color="red", label=r'f(x)= $2x^4 - 3x^2 + 3x - 4$')
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-15, 15)
plt.ylim(-15, 15)
# Mostrarlo.
plt.legend()
plt.show()
```
