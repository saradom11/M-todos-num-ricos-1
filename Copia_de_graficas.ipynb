## GRÁFICA DE √X+1 Y SUS DERIVADAS

```python
import matplotlib.pyplot as plt #Para graficar
import numpy as np #Para hacer caclculos con arreglos
```

```python
# Función cuadrática.
def f(x):
  return (x+1)**(1/2)
def g(x):
  return 1 + (1/2)*x
def h(x):
  return 1 + (1/2)*x -(1/8)*x**(2)
def i(x):
  return 1 + (1/2)*x - (1/8)*x**(2) + (1/16)*x**(3)
```

```python
f(-5)
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace(-10, 10,100)
```

```python
# Graficar ambas funciones.
plt.figure(figsize=(8,8))
plt.title(r'Gráfica de $\sqrt{x+1}$', fontsize=20)
plt.xlabel('Eje x', fontsize=16)
plt.ylabel('Eje y', fontsize=16)
plt.plot(x, f(x), color="yellow", label=r'f(x)= $\sqrt{x+1}$')
plt.plot(x, g(x), color="purple", label=r'g(x)= $1+\frac{1}{2}x$')
plt.plot(x, h(x), color="orange", label=r'h(x)= $1+\frac{1}{2}x-\frac{1}{8}x^2$')
plt.plot(x, i(x), color="blue", label=r'i(x)= $1+\frac{1}{2}x-\frac{1}{8}x^2+\frac{1}{16}x^3$')
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-2, 10)
plt.ylim(-10, 10)
#plt.savefig("output.png") # Guardar gráfico como imágen PNG.
plt.grid(color="pink")
# Mostrarlo.
plt.legend()
plt.show()
```
