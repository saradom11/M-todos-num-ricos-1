# Método de Newton: ejercicio_punto_fijo.png

```python
from math import *
import numpy as np
import matplotlib . pyplot as plt
```

```python
def f(x):
  return x**3-x-1  # Ecuación original
def f_prima(x):
  return 3*x**2-1  # Calculamos la derivada de la ecuación original
```

```python
p0 = 1  # Valor inicial
tolerancia = 0.001  # Tolerancia
max_iter = 100  # Número máximo de iteraciones
```

```python
# Método de Newton

# Imprimir encabezado de la tabla
print("{:^10} | {:^10} | {:^10}".format("Iteración", "p_n", "Error"))
print('----------------------------------------')

# Ciclo iterativo
for i in range(1, max_iter + 1):  # Creamos un ciclo que repita el mismo proceso hasta el número máximo de iteraciones
    p = p0 - ( f(p0) / f_prima(p0) ) # Fórmula para hallar P por el método de Newton
    error = abs(p - p0)  # Calculamos el error |p-p0|

    # Imprimir la iteración, valor de p_n y el error
    print("{:^10} | {:^10.6f} | {:^10.6f}".format(i, p, error))

    # Verificar si el error es menor que la tolerancia y de serlo arrojar la respuesta con los valores obtenidos
    if error < tolerancia:
        print("\nLa solución aproximada es p = {:.6f} con un error de {:.6f}".format(p, error))
        break # Terminar el ciclo para no generer un bucle infinito

    # Actualizar el valor de p0 para la siguiente iteración
    p0 = p
```

```python
# Valores del eje x que toma el gráfico.
x=np.linspace(-100, 100,400)
```

```python
# Gráfica de la función
plt.figure(figsize=(5,5))
plt.title(r'Gráfica de $3x^{2}+1 $', fontsize=20)
plt.xlabel('Eje x', fontsize=14)
plt.ylabel('Eje y', fontsize=14)
plt.plot(x, f_prima(x), color="red", label=r'f´(x)= $3x^{2}+1$')
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid(color="pink")
# Mostrarlo.
plt.legend()
plt.show()
```
