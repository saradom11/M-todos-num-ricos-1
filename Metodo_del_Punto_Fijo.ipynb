# MÃ©todo del Punto Fijoejercicio_punto_fijo.png

```python
from math import *
import numpy as np
import matplotlib . pyplot as plt
```

```python
def f(x):
  return x**3-x-1 #EcuaciÃ³n original
def g(x):
    return (x + 1)**(1/3)  # Primero, reescribimos la ecuaciÃ³n original
```

```python
#Evaluamos en los valores del intervalo para verificar que la nueva ecuaciÃ³n tambien esta contenida
g1 = g(1)
g2 = g(2)

print(f"g(1) = {g1:.4f}")  # Redondeamos a 4 decimales
print(f"g(2) = {g2:.4f}")
```

```python
p0 = 1  # Valor inicial
tol = 0.001  # Tolerancia para el error
nmax_iter = 100  # NÃºmero mÃ¡ximo de iteraciones
```

```python
# MÃ©todo del Punto Fijo

# Primero realizamos el encabezado de la tabla
print("{:^10} | {:^10} | {:^10}".format("IteraciÃ³n", "p_n", "Error"))
print('---------------------------------------------')

# Ciclo iterativo
for i in range(1, nmax_iter + 1):   # Creamos un ciclo que repita el mismo proceso hasta el nÃºmero mÃ¡ximo de iteraciones
    p = g(p0)   # Calcular el siguiente valor p(n) evaluado en la nueva funciÃ³n
    error = abs(p - p0)   # Calcular el error |p-p0|

    # Imprimir la iteraciÃ³n, el valor de p y el error
    print("{:^10} | {:^10.6f} | {:^10.6f}".format(i, p, error))

    # Verificar si el error es menor que la tolerancia y arrojar la respuesta con los valores obtenidos
    if error < tol:
        print("\nLa soluciÃ³n aproximada es p = {:.6f} con un error de {:.6f}".format(p, error))
        break

    # Actualizamos p0 para repetir el ciclo
    p0 = p
```

```python
# Valores del eje x que toma el grÃ¡fico.
x=np.linspace(-100, 100,400)
```

```python
# GrÃ¡fica de la funciÃ³n
plt.figure(figsize=(5,5))
plt.title(r'GrÃ¡fica de $(x+1)^{1/3}$', fontsize=20)
plt.xlabel('Eje x', fontsize=14)
plt.ylabel('Eje y', fontsize=14)
plt.plot(x, g(x), color="red", label=r'g(x)= $(x+1)^{1/3}$')
plt.axhline(1, color="black")
plt.axvline(1, color="black")
plt.axhline(2, color="black")
plt.axvline(2, color="black")
plt.xlim(0.5, 2.5)
plt.ylim(0.5, 2.5)
# Mostrarlo.
plt.legend()
plt.show()
```
