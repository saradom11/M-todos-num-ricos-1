# Gráfica de Maclaurin $f(x)=e^{x/2} * sin (x/3)$

```python
import matplotlib.pyplot as plt #Para graficar
import numpy as np #Para hacer caclculos con arreglos
import sympy as sp #Libreria para calcular derivadas
```

```python
x = sp.symbols('x') #Definir la variable
f = sp.exp(x/2) * sp.sin(x/3) #Colocar la funcion
```

```python
primera_derivada = sp.diff(f, x)  # Primera derivada
segunda_derivada = sp.diff(primera_derivada, x)  # Segunda derivada
tercera_derivada = sp.diff(segunda_derivada, x)  # Tercera derivada
print("Función original:", f)
print("Primera derivada:", primera_derivada)
print("Segunda derivada:", segunda_derivada)
print("Tercera derivada:", tercera_derivada)
```

```python
# Función cuadrática.
def f(x):
  return np.exp(x / 2) * np.sin(x / 3)
def g(x):
  return 1/3*(x)
def h(x):
  return 1/3*(x)+ 1/6*(x)**2
def i(x):
  return 1/3*(x)+ 1/6*(x)**2 + 23/108*(x)**3
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace(-100, 100,400)
```

```python
# Graficar ambas funciones.
plt.figure(figsize=(8,8))
plt.title(r'Gráfica de $e^{x/2} * sin (x/3)$', fontsize=20)
plt.xlabel('Eje x', fontsize=16)
plt.ylabel('Eje y', fontsize=16)
plt.plot(x, f(x), color="yellow", label=r'f(x)= $e^{-x/2} * sin (x/3)$')
plt.plot(x, g(x), color="purple", label=r'g(x)= $1+\frac{1}{3}x$')
plt.plot(x, h(x), color="orange", label=r'h(x)= $1+\frac{1}{3}x+\frac{1}{6}x^2$')
plt.plot(x, i(x), color="blue", label=r'i(x)= $1+\frac{1}{3}x+\frac{1}{6}x^2+\frac{23}{108}x^3$')
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-100, 100)
plt.ylim(-100, 100)
#plt.savefig("output.png") # Guardar gráfico como imágen PNG.
plt.grid(color="pink")
# Mostrarlo.
plt.legend()
plt.show()
```
