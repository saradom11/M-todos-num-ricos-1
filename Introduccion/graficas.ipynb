# Gr치fica de $f(x)=x - 3^{-x}$

```python
import matplotlib.pyplot as plt #Para graficar
import numpy as np #Para hacer caclculos con arreglos
```

```python
# Funci칩n cuadr치tica.
def f(x):
  return x + 3**(-x)
def g(x):
  return x
def h(x):
  return - 3**(-x)
```

```python
f(-5)
```

```python
# Valores del eje X que toma el gr치fico.
x=np.linspace(-10, 10,100)
```

```python
# Graficar ambas funciones.
#plt.plot(x, f(x))
plt.figure(figsize=(10,10))
plt.title('Gr치fica x-3**-x')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.plot(x, g(x), color="purple")
plt.plot(x, h(x), color="orange")
plt.legend({'g(x) = (x)', 'h(x)=3**-x'})
plt.axhline(1, color="red")
plt.axvline(0, color="green")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
#plt.savefig("output.png") # Guardar gr치fico como im치gen PNG.
plt.grid(color="pink")
# Mostrarlo.
plt.show()
```
