```python
# Primero tenemos que importar las librerías:
import numpy as np   # Para poder usar arreglos
import matplotlib.pyplot as plt   # Para poder graficar
from sympy import symbols, simplify  # Para hacer y simplificar el polinomio
```

```python
# Ingresare unos datos de ejemplo:
x = [0, 1, 2, 3]  # Lista con las coordenadas de los puntos "x"
y = [1, 2, 0, 5]  # Lista con las coordenadas de los puntos "y"
```

```python
# Definir la variable simbólica
t = symbols('t')  # Definimos la variable para construir el polinomio

# Construir el polinomio de interpolación de Lagrange
n = len(x)  # Número de puntos en la tabla de datos
polinomio = 0
for i in range(n):
    termino = y[i]  # Inicializamos el término con y[i]
    for j in range(n):
        if i != j:  # Con esto evitamos la división por 0
            termino *= (t - x[j]) / (x[i] - x[j])  # Con la fórmula construimos cada termino
    polinomio += termino  # Añade el término al polinomio
polinomio = simplify(polinomio)
```

```python
# Evaluar el polinomio en un rango de valores para graficar
t_vals = np.linspace(min(x), max(x), 500)  # Genera 500 puntos entre el mínimo y el máximo de x
p_vals = [float(polinomio.subs(t, val)) for val in t_vals]  # Evaluamos el polinomio en los puntos

# Graficar los datos originales y el polinomio
plt.figure(figsize=(9, 5)) #Tamaño de imagen de la gráfica
plt.plot(t_vals, p_vals, label='Polinomio de Lagrange', color='pink')  # Graficamos el polinomio
plt.scatter(x, y, label='Datos originales', color='magenta')  # Ocupamos .scatter para gráficar los puntos
plt.title('Interpolación de Lagrange', fontsize=15) # Titulo de la tabla
plt.xlabel('Eje x', fontsize=12)  # Nombramos el eje 'x'
plt.ylabel('Eje y', fontsize=12)  # Nombramos el eje 'y'
plt.legend()
plt.grid(True)
plt.show()

# Mostrar el polinomio resultante
print("Polinomio de interpolación de Lagrange:")
print(polinomio)
```
