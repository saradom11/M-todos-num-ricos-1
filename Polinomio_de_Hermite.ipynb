# Polinomio de Hermite:

```python
# Primero importamos las librerías
import numpy as np
```

```python
# Datos del ejemplo
valores_x = [1.3, 1.6, 1.9] # Número de puntos dados
valores_fx = [0.6200860, 0.4554022, 0.2818186]
valores_fprimax = [-0.5220232, -0.5698959, -0.5811571]

# Punto que queremos evaluar
evaluación_x = 1.5
```

```python
# Método de Hermite:

# Paso 1: Inicialización de z y Q
n = len(valores_x) # Número de puntos dados
z = [0] * (2 * n) # Lista para duplicar cada x_k porque el método de Hermite usa cada punto dos veces
Q = [[0] * (2 * n) for _ in range(2 * n)] # Una matriz para almacenar las diferencias divididas

# Paso 2: Construcción de z y Q
for i in range(n):
    z[2 * i] = valores_x[i]   # Cada x_k se duplica en z
    z[2 * i + 1] = valores_x[i]
    Q[2 * i][0] = valores_fx[i]  # Las dos primeras filas de Q son f(x_k)
    Q[2 * i + 1][0] = valores_fx[i]  # La derivada f'(x_k) se almacena en la primera columna para los índices impares
    Q[2 * i + 1][1] = valores_fprimax[i]

# Paso 3: Si i≠0
    if i > 0:
        Q[2 * i][1] = (Q[2 * i][0] - Q[2 * i - 1][0]) / (z[2 * i] - z[2 * i - 1])

# Paso 4: Cálculo del resto de Q
for j in range(2, 2 * n):
    for i in range(j, 2 * n):
        Q[i][j] = (Q[i][j - 1] - Q[i - 1][j - 1]) / (z[i] - z[i - j])
```

#### Ocupamos esta fórmula para el paso 4:
image.png

```python
# Evaluación del polinomio en x_eval
resultado = Q[0][0]  # El término independiente
producto = 1.0
for i in range(1, 2 * n):
    producto *= (evaluación_x - z[i - 1])   # Multiplica los factores (x - z[j])
    resultado += Q[i][i] * producto
```

```python
# Resultado
print("Aproximación de f(1.5):", resultado)
```
