# Método del Müller

```python
from math import *
import numpy as np
import matplotlib . pyplot as plt
```

```python
def f(x):
  return x**3 - 13*x - 12 #Ecuación original
```

```python
# Valor inicial
x0 = 4.5
x1 = 5.5
x2 = 5
tolerancia = 0.00001  # Tolerancia para el error
max_iter = 50  # Número máximo de iteraciones
```

```python
# Método de Müller:

# Realizamos el encabezado de la tabla
print("{:^10} | {:^10} | {:^10}".format("Iteración", "x", "Error"))
print('---------------------------------------------')

for i in range(max_iter):  # Creamos un ciclo que repita el mismo proceso hasta el número máximo de iteraciones
    # Valores de la función en x0, x1, x2
    f0 = f(x0)
    f1 = f(x1)
    f2 = f(x2)

    # Diferencias
    h0 = x1 - x0
    h1 = x2 - x1
    d0 = (f1 - f0) / h0
    d1 = (f2 - f1) / h1
    A = (d1 - d0) / (h1 + h0)
    B = A * h1 + d1
    C = f2

    # Determinar el discriminante
    discriminante = np.sqrt(B**2 - 4*A*C)

    # Escoger el denominador más grande para evitar divisiones pequeñas
    if abs(B + discriminante) > abs(B  - discriminante):
        denominador = B + discriminante
    else:
        denominador = B - discriminante

    # Próxima aproximación de la raíz
    dx = -2 * C / denominador
    x3 = x2 + dx

    error = abs(dx)   #Calculamos el error

    # Imprimir la iteración, el valor de p y el error
    print("{:^10} | {:^10.6f} | {:^10.6f}".format(i + 1, x3, error))

    # Verificar si el error es menor que la tolerancia y arrojar la respuesta con los valores obtenidos
    if error < tolerancia:
        print("\nLa solución aproximada es x = {:.6f}".format(x3))
        break

    # Actualizar los valores para la siguiente iteración
    x0, x1, x2 = x1, x2, x3
```

```python
# Valores del eje x que toma el gráfico.
x=np.linspace(-100, 100,400)
```

```python
# Gráfica de la función
plt.figure(figsize=(5,5))
plt.title(r'Gráfica de $x^{3} - 13x - 12$', fontsize=20)
plt.xlabel('Eje x', fontsize=14)
plt.ylabel('Eje y', fontsize=14)
plt.plot(x, f(x), color="red", label=r'f(x)= $x^{3} - 13x - 12$')
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-30, 30)
plt.ylim(-30, 30)
# Mostrarlo.
plt.legend()
plt.show()
```
