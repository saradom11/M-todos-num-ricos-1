# Método de Neville:
### Escribir un programa en Python que implemente el método de Neville para calcular el polinomio interpolante de Lagrange en un punto dado.

```python
# Primero importamos la librería
import numpy as np
```

```python
# Datos de la tabla 3.5
valores_x = np.array([1.0, 1.3, 1.6, 1.9, 2.2])
valores_fx = np.array([0.7651977, 0.6200860, 0.4554022, 0.2818186, 0.1103623])

# Punto que queremos interpolar
evaluacion_x = 1.5
```

##### Fórmula de Neville
image.png

```python
# Implementación del Método de Neville
n = len(valores_x)
Q = np.zeros((n, n))  # Creamos una matriz para almacenar los resultados intermedios de la tabla

# Hacemos la primera columna con los valores de f(x)
for i in range(n):
    Q[i, 0] = valores_fx[i]

# Llenar la tabla de Neville
for j in range(1, n):  # Columna
    for i in range(j, n):  # Fila
        Q[i, j] = ((evaluacion_x - valores_x[i - j]) * Q[i, j - 1] - (evaluacion_x - valores_x[i]) * Q[i - 1, j - 1]) / (valores_x[i] - valores_x[i - j]) # El doble bucle recorre las columnas "j" y las filas "i" para llenar los valores de la tabla usando la fórmula de Neville
        # Sustituimos en un inicio el valor a interpolar en la fórmula (x por 1.5)

# El valor interpolado está en Q[n-1, n-1] (dado la matriz)
result = Q[n - 1, n - 1]

# Imprimimos la tabla de Neville
print("Tabla de Neville:")
print()
print (Q)
print ()

# Imprimimos el resultado interpolado con el resultado de la tabla de Neville
print("El valor interpolado en x =", evaluacion_x, "es f(x) =", result)
```
